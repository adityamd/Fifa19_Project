```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data2 = data.drop(["Unnamed: 0","ID"],axis=1)
data2.head()
```

```python
data2.describe()
```

```python
#Treating NaN Values
for i in data2:
    if data2[i].dtype=='object':
        data2[i]=data2[i].fillna(str(data2[i].mode()[0]))
    else:
        data2[i]=data2[i].fillna(data2[i].mean())
```

```python
def Scaling(data):
    return data/data.max()
```

```python
#Looking For Columns That have URL as values
import re
for i in data2:
        if len(re.findall('http.*',str(data.loc[0,i])))>0:
                print(i)
```

```python
data2=data2.drop(['Photo','Flag','Club Logo'],axis=1)
```

```python
data2.head()
```

```python
#Converting columns with  format '€...M or €...K' from string to numerical with 'K' or 1000 units 
for i in data2:
    if len(re.findall('.[0-9][A-Z]$',str(data2.loc[0,i])))>0:
        for j in data2.index:
            a=str(data2.loc[j,i])[1:]
            if a=='0':
                data2.loc[j,i]=float(a)
                continue
            a=re.findall('(.*)[A-Z]$',a)
            a=''.join(a)
            if str(data2.loc[j,i])[-1]=='M':
                data2.loc[j,i]=str(float(a)*1000)
            else:
                data2.loc[j,i]=float(a)
        data2[i]=data2[i].astype('float64')
```

```python
data2.head()
```

```python
features=data2.loc[:,'Age':'GKReflexes']
```

```python
target=data2.iloc[:,-1]
```

```python
correlation_mat=features.join(target).corr()
high_correlation_mat=correlation_mat[abs(correlation_mat['Release Clause'])>0.3]
features2=features.loc[:,high_correlation_mat.index[:-1]]
```

```python
train_data=np.load('Training_Data.npz')
```

```python
val_data=np.load('Validation_Data.npz')
```

```python
input_size_=train_data['inputs'].shape[1]
hidden_layer_size=15
output_size=1
```

```python
import tensorflow as tf
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(hidden_layer_size),
    tf.keras.layers.Dense(hidden_layer_size,activation='linear',
                        kernel_initializer = tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size))),
                        bias_initializer =  tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size)))),
    tf.keras.layers.Dense(hidden_layer_size, activation='linear',
                        kernel_initializer = tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size))),
                        bias_initializer =  tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size)))),
    tf.keras.layers.Dense(hidden_layer_size, activation='linear',
                        kernel_initializer = tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size))),
                        bias_initializer =  tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size)))),
    tf.keras.layers.Dense(hidden_layer_size, activation='linear',
                        kernel_initializer = tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size))),
                        bias_initializer =  tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size)))),
    tf.keras.layers.Dense(output_size, activation='linear',
                        kernel_initializer = tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size))),
                        bias_initializer =  tf.random_uniform_initializer(np.sqrt(6/(input_size_+output_size))))
])
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5)
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mae'])
```

```python
batch_size_=int(train_data['inputs'].shape[0]/10)
model.fit(train_data['inputs'],train_data['target'],batch_size=batch_size_,callbacks=[callback],epochs=200,validation_data=(val_data['inputs'],val_data['target']),verbose=2,validation_steps=1)
```

```python
test_data=np.load('Testing_Data.npz')
```

```python
predictions=model.predict(test_data['inputs'])
```

```python
rss=((predictions.flatten()-test_data['target'])**2).sum()
tss=((test_data['target']-test_data['target'].mean())**2).sum()
r2=1-(rss/tss)
r2
```
